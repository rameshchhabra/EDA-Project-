```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# CENTURY SCORE SUMMARY 
sns.boxplot(x='Score', data=data_file);
```

```python
#1 NUMBER OF CENTURIES IN EACH BATTING ORDER 
score_data=data_file.groupby(['Batting Order'],group_keys=True)['Batting Order'].count()
print(score_data)
score_data[score_data.values==score_data.max()]


```

```python
#AVERAGE SCORE IN DIFFERENT BATTING ORDER 
score_av=data_file.groupby(['Batting Order'])['Score']
avg=score_av.agg('mean')
avg
```

```python
# NUMBER OF CENTURIES AS CAPTIAN AND AS A PLAYER 
captain=data_file.groupby(['Captain'])['Captain']
captain.count()
```

```python
# NUMBER OF CENTURY AWAY AND HOME
away=data_file.groupby(['H/A'])['H/A']
away.count()

```

```python
#  AVERAGE CENTURIES AVERAGE AGAINST DIFFERENT TEAMS 
country=data_file.groupby(['Against'],as_index=False)['Score']
a=country.agg('mean')
a
a = a.rename(columns={a.columns[1]: 'avg_score'})

#make graph of this data set
sns.lineplot(x="Against", y="avg_score",data=a)
```

```python
#CENTURY CONTRIBUTION IN RESULTS

win=data_file.groupby(['Result'])['Result']
win.count()
```

```python
# NUMBER OF CENTURIES IN EACH YEAR

data=data_file['Date']

dates=pd.to_datetime(data.values,infer_datetime_format=True)
dates

year = dates.strftime("%Y")
year=year.sort_values()

year_data=pd.DataFrame(year ,columns=["year of century"])
counts=year_data.groupby(['year of century'])['year of century'].count()
print(counts)
sns.barplot(x=counts.index ,y=counts)
```

```python
# NO OF CENTURIES AGAINST DIFFERENT TEAMS 
teams=data_file.groupby(['Against'])['Score'].count()
print(teams)
```

```python
#PIE CHART TO SHOW PERCENTAGE SHARE OF NO OF CENTURY AGAINST DIFFERENT TEAMS 

d=data_file['Against'].value_counts()
d.plot.pie(autopct='%1.2f%%')
```

```python
#BAR PLOT TO SHOW COUNT OF NO OF CENTURIES IN DIFFERENT VENUES 

a=data_file['Venue'].value_counts().sort_index(ascending=True)
a.plot.bar(color=['black', 'red', 'green', 'blue'])
```

```python
#FINDING MOST NO OF CENTURIES AGAINST A PARTICULAR TEAM 

a = data_file['Against'].value_counts()
most_century_against = a.iloc[:1]
most_century_against
```
